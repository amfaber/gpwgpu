```python
bounds = np.array([7, 10])
coords = np.array([1, 4])
N = len(bounds)

strides = np.zeros(N)
next_stride = 1
for i in range(N):
    strides[N - 1 - i] = next_stride
    next_stride *= bounds[N - 1 - i]


new_strides = np.zeros(N)
new_coords = np.zeros(N)
next_stride = 1
new_idx = 0
for i in range(N - 1):
    new_strides[N - 1 - i] = next_stride
    new_idx += next_stride * coords[N - 2 - i]
    next_stride *= bounds[N - 2 - i]
    new_coords[N - 1 - i] = coords[N - 2 - i]
new_strides[0] = next_stride
new_coords[0] = coords[N - 1]
new_idx += next_stride * coords[N - 1]


idx = sum(coords * strides)
print(idx)
print(new_idx)
```
